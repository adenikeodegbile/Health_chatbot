```python
!pip install numpy==1.24.4
import os
import torch
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader, ServiceContext
# Import the Hugging Face LLM integration
from llama_index.llms.huggingface import HuggingFaceLLM
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
```

```python
# Define the model to use from Hugging Face.
# BioMistral-7B is a strong choice for a medical project.
model_id = "BioMistral/BioMistral-7B"
```

```python
# Set up quantization for more efficient memory usage
# This is recommended if you have limited GPU VRAM
quantization_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)
```

```python
# Load the tokenizer and the model from Hugging Face Hub
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    quantization_config=quantization_config,
    device_map="auto" # This automatically detects and uses your GPU
)
```
